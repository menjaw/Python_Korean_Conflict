```python
#Download data from URL
import webget
import pandas as pd


url = 'https://raw.githubusercontent.com/PatrickFenger/pythonAssignments/master/KoreanConflict.csv'

response = webget.download(url)
print("File downloadet", response)
```

```python
#Read file
dataframe = pd.read_csv("KoreanConflict.csv")
#Print all content of file by removing # from print 
#print(dataframe)
```

```python
#Q1: How many soldiers entered from a marine corps branch?
branch = dataframe['BRANCH']=='MARINE CORPS'

print(dataframe.groupby(branch).size())
print("ANSWER: 4509 soldiers entered from a Marine Corps")
```

```python
# Q2: Which enrollment was the most common?
enrollment = dataframe['ENROLLMENT']

print(dataframe.groupby(enrollment).size())
print("ANSWER: The most common enrollment was ACTIVE - REGULAR with 23973")
```

```python
# Q3: Was there an ethnicity majority throughout the war - If so, which?
et1 = dataframe['ETHNICITY']
et2 = dataframe['ETHNICITY_1']
et3 = dataframe['ETHNICITY_2']

print(dataframe.groupby(et1).size())
print(dataframe.groupby(et2).size())
print(dataframe.groupby(et3).size())

print("ANSWER: Yes there is an ethnicity majority: White")
```

```python
#Q4: Which division had the most casualties(tilskadekomne)?
division = dataframe['DIVISION'] #This is the row
fatality = dataframe['FATALITY'] #This is the column

print(pd.crosstab(division, fatality))
print('ANSWER: 1 MAR DIV had most casualties')
```

```python
# Q5: Which Home state suffered the most losses?
state = dataframe['HOME_STATE']
print(dataframe.groupby(state).size())
print('ANSWER: California with 2582 losses')
```
